**Here is an example of the data format:**
### Business Objects

Business objects contain basic information about local businesses. The fields are as follows:

```json
{
  'type': 'business',
  'business_id': (a unique identifier for this business),
  'name': (the full business name),
  'neighborhoods': (a list of neighborhood names, might be empty),
  'full_address': (localized address),
  'city': (city),
  'state': (state),
  'latitude': (latitude),
  'longitude': (longitude),
  'stars': (star rating, rounded to half-stars),
  'review_count': (review count),
  'photo_url': (photo url),
  'categories': [(localized category names)]
  'open': (is the business still open for business?),
  'schools': (nearby universities),
  'url': (yelp url)
}
```
### Checkin Objects
```json
{
    'type': 'checkin',
    'business_id': (encrypted business id),
    'checkin_info': {
        '0-0': (number of checkins from 00:00 to 01:00 on all Sundays),
        '1-0': (number of checkins from 01:00 to 02:00 on all Sundays),
        ...
        '14-4': (number of checkins from 14:00 to 15:00 on all Thursdays),
        ...
        '23-6': (number of checkins from 23:00 to 00:00 on all Saturdays)
    }, # if there was no checkin for a hour-day block it will not be in the dict
}
```

# Data Collection/Processing: 

````python
import json
import pandas as pd
business_json_path = 'yelp_academic_dataset_business.json'
df = pd.read_json(business_json_path, lines=True)


# Here We are creating an extra column to the database which helps to predict the star score better.
attScore = []
for i in df['attributes']:
    s = 0
    if i != None:
        for k,v in i.items():
            if v == "True":
                s = s + 1
    attScore = attScore + [s] 
df['Attribute_Score'] = attScore
````

````python
# Displaying the total length of the dataset
print(len(df))


#Removing the Null Values as part of the data preprocessig step.
df = df.dropna()

# Again displaying the values after the null values are removed.
print(len(df))
````

````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# this step is performed to normalize the dataset we have for better results.

# creating the MinMaxScaler object
scaler = MinMaxScaler()

# selecting only the columns which we will be using to normalize
columns_to_normalize = ['latitude', 'longitude', 'review_count','Attribute_Score' ]

# normalize the selected columns
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
````

# Solution : 

We have a dataset that contains information about businesses, including their category, location, and review count. Our target variable is the "stars" column, which ranges from 1 to 5 (in half-star increments).
The dataset includes both numerical and categorical variables, which will require different preprocessing techniques. We will need to handle missing data, outliers, and categorical variables with a large number of unique values.
The success of this project could benefit both consumers and business owners. Consumers can use this model to find businesses that are likely to meet their expectations, while business owners can use it to identify areas for improvement and increase their overall rating.

I have used 3 Machine learning models to predict the star rating of the business 
1.Multiple Linear regression
2. Random Forest Regressor
3. KNN


Code to implement the solution in python:

````python
# Code to implement the Multiple Linear Regression Model

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Selecting the features we want to use for prediction
features = ['latitude', 'longitude', 'review_count', 'Attribute_Score']

# Extracting the features and target variable from the dataframe
X = df[features] # predictors
y = df['stars']  # Target

# Splitting the data into training and testing sets with 25% 75% split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create a linear regression model and fit it to the training data
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the star rating for the testing data
y_pred_linear = model.predict(X_test)

# Calculate the mean squared error (MSE) of the predictions
mse = mean_squared_error(y_test, y_pred_linear)

mse_linear = mse
y_test_linear = y_test

print('Mean Squared Error:', mse)
````

````python
# Code to implement the RandomForestRegressor

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error



# Selecting relevant features
features = ['latitude', 'longitude', 'review_count', 'Attribute_Score']

# Convert categorical features to numerical
#data = pd.get_dummies(data, columns=['attributes'])#, 'BusinessAcceptsCreditCards', 'NoiseLevel', 'OutdoorSeating', 'RestaurantsAttire', 'RestaurantsDelivery', 'RestaurantsReservations', 'RestaurantsTakeOut'])


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df['stars'], test_size=0.2, random_state=42)

# Train a random forest regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
y_pred = rf.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean squared error: {:.2f}'.format(mse))

mse_forest = mse

y_pred_forest = y_pred
y_test_forest = y_test
````

````python
# Code to implement the K Nearest Neighbours

import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

def knn(k):
    # defining the relevant features for predicting the rating
    X = df[['review_count', 'latitude', 'longitude', 'Attribute_Score']]
    y = df['stars']

    # Splitting the dataset into training and testing sets 30% and 70%
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Train the KNN model
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Predict the ratings for the test set
    y_pred = knn.predict(X_test)

    # Evaluate the performance of the model using Mean Squared Error and R2 Score
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    #print("Mean Squared Error:", mse)
    
    return mse

````

````python

mknn = []
kvalues= [5,10,16,19,25,30,41,56,77,89,100,150, 170, 200, 250]
for i in kvalues:
    ab = knn(i)
    mknn = mknn + [ab]
#print(mknn)


import matplotlib.pyplot as plt
plt.figure(figsize=(25,10))
plt.plot(kvalues, mknn,color = 'red')

plt.title("MSE analysis based on K values ",fontdict={'fontsize': 20})
plt.xlabel('K Values',fontdict={'fontsize': 15})
plt.ylabel('MSE Values',fontdict={'fontsize': 15})
plt.legend()
plt.show()
````

````python
# Code to implement the K Nearest Neighbours

#based on the graph above selecting the k valye as 75

import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


# defining the relevant features for predicting the rating
X = df[['review_count', 'latitude', 'longitude', 'Attribute_Score']]
y = df['stars']

# Splitting the dataset into training and testing sets 30% and 70%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the KNN model
knn = KNeighborsRegressor(n_neighbors=70)
knn.fit(X_train, y_train)

# Predict the ratings for the test set
y_pred = knn.predict(X_test)

# Evaluate the performance of the model using Mean Squared Error and R2 Score
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)


mse_knn = mse
y_pred_knn = y_pred
y_test_knn = y_test
````
