**Here is an example of the data format:**
### Business Objects

Business objects contain basic information about local businesses. The fields are as follows:

```json
{
  'type': 'business',
  'business_id': (a unique identifier for this business),
  'name': (the full business name),
  'neighborhoods': (a list of neighborhood names, might be empty),
  'full_address': (localized address),
  'city': (city),
  'state': (state),
  'latitude': (latitude),
  'longitude': (longitude),
  'stars': (star rating, rounded to half-stars),
  'review_count': (review count),
  'photo_url': (photo url),
  'categories': [(localized category names)]
  'open': (is the business still open for business?),
  'schools': (nearby universities),
  'url': (yelp url)
}
```
### Checkin Objects
```json
{
    'type': 'checkin',
    'business_id': (encrypted business id),
    'checkin_info': {
        '0-0': (number of checkins from 00:00 to 01:00 on all Sundays),
        '1-0': (number of checkins from 01:00 to 02:00 on all Sundays),
        ...
        '14-4': (number of checkins from 14:00 to 15:00 on all Thursdays),
        ...
        '23-6': (number of checkins from 23:00 to 00:00 on all Saturdays)
    }, # if there was no checkin for a hour-day block it will not be in the dict
}
```

# Data Collection/Processing: 

````python
#----------------------------------------------
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

import json
import pandas as pd
business_json_path = 'yelp_academic_dataset_business.json'
df = pd.read_json(business_json_path, lines=True)


# Here We are creating an extra column to the database which helps to predict the star score better.
attScore = []
for i in df['attributes']:
    s = 0
    if i != None:
        for k,v in i.items():
            if v == "True":
                s = s + 1
    attScore = attScore + [s] 
df['Attribute_Score'] = attScore
````

````python
df
````

````python
# Displaying the total length of the dataset
print(len(df))


#Removing the Null Values as part of the data preprocessig step.
df = df.dropna()

# Again displaying the values after the null values are removed.
print(len(df))
````

````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# this step is performed to normalize the dataset we have for better results.

# creating the MinMaxScaler object
scaler = MinMaxScaler()

# selecting only the columns which we will be using to normalize
columns_to_normalize = ['latitude', 'longitude', 'review_count','Attribute_Score' ]

# normalize the selected columns
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
````

# Data Exploration: Exploring the Yelp Dataset

**(1) Finding the most popular business categories:** 
* print the top 10 most popular business categories in the dataset and their counts in a table (i.e., how many business objects in each category). Here we say a category is "popular" if there are many business objects in this category (such as 'restaurants').

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
from collections import Counter
from prettytable import PrettyTable


c = ""
for j in df['categories']:
    if j != None:
        c = c + j
categories = c.split(",")

counts = Counter(categories)
a = list(counts.most_common(10))

xa = PrettyTable()

xa.field_names = ["S.No","Business Category", "count"]


for i in range(10):
    #print(bussiness_id[i], checkins[i])
    xa.add_row([i+1,a[i][0], a[i][1]])
print(xa)
````

**(2) Find the most popular business objects** 
* print the top 10 most popular business objects/IDs in the dataset and their counts (i.e., how many checkins in total for each business object).  Here we say a business object is "popular" if the business object attracts a large number of checkins from the users.

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

import json
import pandas as pd
business_json_path = 'yelp_academic_dataset_checkin.json'
dfc = pd.read_json(business_json_path, lines=True)

````

````python
pop = dict()
for i in range(len(dfc)):
    totalCheck = 0
    for j in dfc['checkin_info'][i].values():
        #print(j)
        totalCheck = totalCheck + j
    pop[dfc['business_id'][i]] =  totalCheck

````

````python


from prettytable import PrettyTable

x = PrettyTable()

x.field_names = ["S.No","Business ID", "Checkins"]
# sort the dictionary by values
sorted_dict = dict(sorted(pop.items(), key=lambda item: item[1], reverse=True))
#print(len(dfc))
#print(len(pop))
# print the sorted dictionary
#print(sorted_dict)
#print(len(sorted_dict))

checkins = list(sorted_dict.values())

bussiness_id = list(sorted_dict.keys())

#print(bussiness_id[:10])
#print(checkins[:10])

for i in range(10):
    #print(bussiness_id[i], checkins[i])
    x.add_row([i+1,bussiness_id[i], checkins[i]])
print(x)
````

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

# Writing a code to show the top 5 Categories with highest rating and lowest ratings

import pandas as pd
import matplotlib.pyplot as plt

business_json_path = 'yelp_academic_dataset_business.json'
yelp_data = pd.read_json(business_json_path, lines=True)

# Select the columns we need
review_data = yelp_data[['business_id', 'stars', 'categories']]

# Calculating the average star rating for each category
category_average_ratings = review_data.groupby('categories', as_index=False)['stars'].mean()
category_average_ratings = category_average_ratings.sort_values(by='stars', ascending=False)

print("Top 5 Categories with the Highest Average Star Rating")
print(category_average_ratings.head(5))

r = range(1,6)

# Ploting the top 5 categories with the highest average star rating
fig, ax = plt.subplots(figsize=(12, 6))
top_categories = category_average_ratings.head(5)

ax.bar(top_categories['categories'], top_categories['stars'])
ax.set_xlabel('Category')
ax.set_ylabel('Average Star Rating')
ax.set_xticks(r)
ax.set_title('Top 5 Categories with the Highest Average Star Rating')
plt.show()


print("Top 5 Categories with the Highest Average Star Rating")
print(category_average_ratings.tail(5))

# Ploting the top 5 categories with the lowest average star rating
fig, ax = plt.subplots(figsize=(12, 6))
bottom_categories = category_average_ratings.tail(5)
ax.bar(bottom_categories['categories'], bottom_categories['stars'])

ax.set_xlabel('Category')
ax.set_ylabel('Average Star Rating')

ax.set_xticks(r)
ax.set_title('Top 5 Categories with the Lowest Average Star Rating')
plt.show()




````

# The Solution: implement a data science solution to the problem you are trying to solve.

````python
# Code to implement the Multiple Linear Regression Model

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Selecting the features we want to use for prediction
features = ['latitude', 'longitude', 'review_count', 'Attribute_Score']

# Extracting the features and target variable from the dataframe
X = df[features] # predictors
y = df['stars']  # Target

# Splitting the data into training and testing sets with 25% 75% split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create a linear regression model and fit it to the training data
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the star rating for the testing data
y_pred_linear = model.predict(X_test)

# Calculate the mean squared error (MSE) of the predictions
mse = mean_squared_error(y_test, y_pred_linear)

mse_linear = mse
y_test_linear = y_test

print('Mean Squared Error:', mse)
````

````python
# Code to implement the RandomForestRegressor

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error



# Selecting relevant features
features = ['latitude', 'longitude', 'review_count', 'Attribute_Score']

# Convert categorical features to numerical
#data = pd.get_dummies(data, columns=['attributes'])#, 'BusinessAcceptsCreditCards', 'NoiseLevel', 'OutdoorSeating', 'RestaurantsAttire', 'RestaurantsDelivery', 'RestaurantsReservations', 'RestaurantsTakeOut'])


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df['stars'], test_size=0.2, random_state=42)

# Train a random forest regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
y_pred = rf.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean squared error: {:.2f}'.format(mse))

mse_forest = mse

y_pred_forest = y_pred
y_test_forest = y_test
````

````python
# Code to implement the K Nearest Neighbours

import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


# defining the relevant features for predicting the rating
X = df[['review_count', 'latitude', 'longitude', 'Attribute_Score']]
y = df['stars']

# Splitting the dataset into training and testing sets 30% and 70%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the KNN model
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict the ratings for the test set
y_pred = knn.predict(X_test)

# Evaluate the performance of the model using Mean Squared Error and R2 Score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R2 Score:", r2)


mse_knn = mse
y_pred_knn = y_pred
y_test_knn = y_test
````
