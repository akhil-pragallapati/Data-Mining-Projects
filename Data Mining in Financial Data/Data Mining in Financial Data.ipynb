# Solution


The solution for this problem can be solved by analysing Patterns and trends that can be discovered by examining historical stock prices, and machine learning models can then be trained to forecast future stock prices.
The objective would be to create a model that correctly forecasts stock prices so that it may be used to guide investing choices.
This can be accomplished by including a variety of elements as the predictors and try to find the target value, which in our case is the Close price of the specific stock.
I have used two machine learning models for the analysis, these are 
    1. Linear regression 
    2. Support Vector Regression(SVR)
   
These models were chosen because regression techniques perform good when predicting discrete values, which is the aim that we are trying to forecast. 


**Write codes to implement the solution in python:**

```python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

# Defining the target variable and features

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
def linearRrgressionStock(tesla, port):
    y = tesla["Close"]
    X = tesla.drop(["Close", "Date"], axis=1)



    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)

    reg = LinearRegression().fit(X_train, y_train)

    # Make predictions on the test set
    predictions = reg.predict(X_test)
    plt.figure(figsize=(25,10))
    # Plotting the predictions against the actual stock prices
    plt.plot(predictions, label="Predicted stock price")
    plt.plot(y_test.values, label="Actual stock price")
    plt.xlabel("Days",fontdict={'fontsize': 15})
    plt.ylabel("Closing price of " + port,fontdict={'fontsize': 15})
    plt.title(port + " Linear Regression Model ",fontdict={'fontsize': 20} )
    plt.legend()
    plt.show()

    # Calculate the mean squared error
    mse = np.mean((predictions - y_test) ** 2)
    print("Mean Squared Error:", mse)
    return mse

```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier

def SVRStock(tesla, port):
    y = tesla["Close"]
    X = tesla.drop(["Close", "Date"], axis=1)



    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)

    reg = SVR().fit(X_train, y_train)

    # Make predictions on the test set
    predictions = reg.predict(X_test)
    plt.figure(figsize=(25,10))
    # Plotting the predictions against the actual stock prices
    plt.plot(predictions, label="Predicted stock price")
    plt.plot(y_test.values, label="Actual stock price")
    plt.xlabel("Days",fontdict={'fontsize': 15})
    plt.ylabel("Closing price of " + port,fontdict={'fontsize': 15})
    plt.title(port + " SVR Model " ,fontdict={'fontsize': 20} )
    plt.legend()
    plt.show()

    # Calculate the mean squared error
    mse = np.mean((predictions - y_test) ** 2)
    print("Mean Squared Error:", mse)
    return mse
```
